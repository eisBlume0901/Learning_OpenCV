```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("apollo_saturn_v_rocket.jpg", cv2.IMREAD_UNCHANGED)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_line = img.copy()

cv2.line(img_line, (200, 100), (400, 100), (255, 255, 0), thickness=5, lineType=cv2.LINE_AA)
plt.imshow(img_line)
```

```python
img_circle = img.copy()

cv2.circle(img_circle, (200, 300), 50, (255, 0, 0), thickness=3, lineType=cv2.LINE_AA)

plt.imshow(img_circle)
```

```python
img_rectangle = img.copy()

cv2.rectangle(img_rectangle, (250, 300), (350, 150), (0, 0, 255), thickness=3, lineType=cv2.LINE_8)

plt.imshow(img_rectangle)
```
